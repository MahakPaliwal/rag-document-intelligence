```python
import fitz  # PyMuPDF

def load_pdf(file_path):
    doc = fitz.open(file_path)
    full_text = ""
    for page in doc:
        full_text += page.get_text()
    return full_text

# Load your uploaded PDF
text = load_pdf(file_name)

# Check it worked
print(f"Total characters extracted: {len(text)}")
print("\nFirst 500 characters preview:")
print(text[:500])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

def split_text(text):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,     # each chunk = ~500 characters
        chunk_overlap=50    # 50 char overlap between chunks
    )
    chunks = splitter.split_text(text)
    return chunks

chunks = split_text(text)
print(f"Total chunks created: {len(chunks)}")
```

```python
for i, chunk in enumerate(chunks):
    print(f"\n--- Chunk {i+1} ---")
    print(chunk)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)
print("Embedding model loaded!")
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_texts(chunks, embeddings)
print("Vectorstore created!")
print(f"Total vectors stored: {vectorstore.index.ntotal}")
```

```python
query = "What is this paper about?"
results = vectorstore.similarity_search(query, k=3)

print(f"Top 3 relevant chunks for: '{query}'\n")
for i, doc in enumerate(results):
    print(f"--- Result {i+1} ---")
    print(doc.page_content)
    print()
```

```python
from groq import Groq

client = Groq()

def ask_question(question, vectorstore):
    # Retrieve relevant chunks
    docs = vectorstore.similarity_search(question, k=3)
    context = "\n\n".join([d.page_content for d in docs])

    # Build prompt
    prompt = f"""Use only the context below to answer the question.
If the answer is not in the context, say "I don't know."

Context:
{context}

Question: {question}
Answer:"""

    # Send to Groq LLM
    response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",  # ← changed this line only
    messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

# Test it
answer = ask_question("What is this paper about?", vectorstore)
print(answer)
```

```python
questions = [
    "What method is used for fingerprint restoration?",
    "What metrics are used to evaluate the model?",
    "What are the applications of this research?"
]

for q in questions:
    print(f"Q: {q}")
    print(f"A: {ask_question(q, vectorstore)}")
    print()
```

```python
from IPython.display import Markdown, display

print("RAG System Ready! Type your question below.")
print("Type 'exit' to stop.\n")

while True:
    question = input("Your question: ")

    if question.lower() == "exit":
        print("Goodbye!")
        break

    if question.strip() == "":
        print("Please enter a valid question.\n")
        continue

    answer = ask_question(question, vectorstore)
    display(Markdown(f"**Answer:** {answer}"))
    print()
```
